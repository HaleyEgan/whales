```python
#merge coords and data tables on 'Country Code'
#df1.merge(df2, on="column_name", how="left")
#coords.join(data, on='Country Code', how='right')
#pd.merge(data, coords, on='Country Code', how='left')
#data = data.merge(coord, on='Country Code', how='left')
#data.join(coords, on='Country Code', how='left')
#new = data.merge(coords, on='Country Code', how='left')
#new = data.join(coords.set_index('Country Code'), on = 'Country Code', how = 'left')
data = pd.merge(data, coords, on='Country Code', how='left')
```

```python
data.head()
```

```python
#find all unique values for Port Size
data['Port Size'].unique()
```

#### Groupings
- Very Small & Small
- Medium
- Large & Very Large

```python
#find all unique values for Port Type
data['Port Type'].unique()
```

```python
#sort out largest ports - large and very large
#largest_ports = data.loc[data['Port Size'] == 'Large']
largest_ports = data.loc[data['Port Size'].isin(['Large', 'Very Large'])]
```

```python
largest_ports.sample(10)
```

```python
len(largest_ports)
```

```python
#sort out smallest ports - small and very small
smallest_ports = data.loc[data['Port Size'].isin(['Small', 'Very Small'])]
smallest_ports.sample(10)
```

```python
len(smallest_ports)
```

```python
#sort out mid-size ports 
medium_ports = data.loc[data['Port Size'] == 'Medium']
medium_ports.sample(10)
```

```python
len(medium_ports)
```

```python
#save new tables as csv files
largest_ports.to_csv("LargePorts.csv", index=False)
medium_ports.to_csv("MediumPorts.csv", index=False)
smallest_ports.to_csv("SmallPorts.csv", index=False)
```

#### Groupings
- Likely coastal and large vessels
    - Seaport - Deepwater Seaport - Port Terminal - Off-Shore Terminal 
- Other
    - River Port - Canal - Waterway Region - Ferry - Pier, Jetty or Wharf - Marina - Harbor (?) 

```python
#Mostly Coastal ports (based on port type description)
Coastal_ports = data.loc[data['Port Type'].isin(['Seaport', 'Deepwater Seaport', 'Port Terminal', 'Off-Shore Terminal', 'Ferry', 'Pier, Jetty or Wharf'])]
print(len(Coastal_ports))
Coastal_ports.sample(10)
```

```python
#Other ports (based on port type description)
Other_ports = data.loc[data['Port Type'].isin(['River Port', 'Canal', 'Waterway Region', 'Marina', 'Harbor'])]
print(len(Other_ports))
Other_ports.sample(10)
```

```python
Coastal_ports.to_csv("CoastalPorts.csv", index=False)
Other_ports.to_csv("OtherPorts.csv", index=False)
```
